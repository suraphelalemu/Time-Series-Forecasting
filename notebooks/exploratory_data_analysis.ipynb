```python

import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  
import os  
import sys
```

```python


# 🛠️ Add the 'scripts' directory to the Python path for module imports
sys.path.append(os.path.abspath(os.path.join('..', 'scripts')))

```

```python


# 📚 Importing the DataPreprocessor class from the scripts directory
from data_preprocessing import DataPreprocessor 
from exploratory_data_analysis import StockDataExploration

```

```python


# 🛠️ Create an instance of DataPreprocessor
processor = DataPreprocessor()
analyzer = StockDataExploration()

```

```python


bnd_data = processor.load_data('BND')
spy_data = processor.load_data('SPY')
tsla_data = processor.load_data('TSLA')

```

```python
# Define dictionary for stock data for TSLA, BND, SPY
data_dict = {
    'BND': bnd_data,    
    'SPY': spy_data,  
    'TSLA': tsla_data 
}

# Print statement with emojis for better visualization
print("📈 Stock Data Dictionary Created Successfully! 🚀")
print("Stocks Included:")
for symbol in data_dict.keys():
    print(f" ✅ {symbol} 📊")
```

```python


# 📈 Plotting the closing price over time 
print("📊 Preparing to Plot Closing Prices Over Time... ⏳")
analyzer.plot_closing_price(data_dict)
print("✅ Closing Prices Plotted Successfully! 🎉")

```

```python


# 📈 Plotting the daily percentage change in the closing price to observe volatility.
print("📊 Preparing to Plot Daily Percentage Change in Closing Prices... ⏳")
analyzer.plot_percentage_change(data_dict)
print("✅ Daily Percentage Change Plotted Successfully! 🎉")

```

```python
# 📈 Plotting the closing price, rolling mean, and volatility (rolling std) over time for multiple symbols.
print("📊 Starting Analysis of Price Trends... ⏳")
analyzer.analyze_price_trend(data_dict)
print("✅ Price Trend Analysis Completed Successfully! 🎉")
```

```python


# 🔍 Detecting outliers in the stock data
print("📊 Initiating Outlier Detection... ⏳")
analyzer.detect_outliers(data_dict)
print("✅ Outlier Detection Completed Successfully! 🎉")

```

```python


# 📈 Plotting the closing price for a stock symbol (e.g., 'TSLA')
stock_symbol = 'TSLA'  # Specify the stock symbol you want to plot
print(f"📊 Preparing to Plot Daily Percentage Change for {stock_symbol}... ⏳")
analyzer.plot_percentage_change({stock_symbol: data_dict[stock_symbol]})
print("✅ Daily Percentage Change Plotted Successfully! 🎉")

```

```python


# 📈 Plotting unusual daily returns for the stock data
print("📊 Initiating Plot of Unusual Daily Returns... ⏳")
analyzer.plot_unusual_daily_return(data_dict)
print("✅ Unusual Daily Returns Plotted Successfully! 🎉")

```

```python


# 📊 Analyzing trends and seasonality in the stock data
print("📈 Starting Analysis of Trends and Seasonality... ⏳")
analyzer.analyze_trends_and_seasonality(data_dict)
print("✅ Trends and Seasonality Analysis Completed Successfully! 🎉")

```
